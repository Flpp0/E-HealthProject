```python
df_Complete = df.copy()

## Removed gender_2, gender_3, 'EHEALS_sum', 'GAD_sum', 'PHQ_sum'
df = df.loc[:,['age','HEALS_sum','CCS_sum','PHQ_sum','GAD_sum','EHEALS_sum']]
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

X_min = 0
X_max_PHQ = 27
X_max_GAD = 21
X_max_EHEALS = 40
X_max_HEALS = 39 ## I know that is X_max_HEAS !!!!!!
X_max_CCS = 72

# Extracting the numerical variables: in our case are the ones that non startswith 'gender'
numerical_columns = df.columns[~df.columns.str.startswith('gender_')]

numerical_data = df[numerical_columns]

#scaler = MinMaxScaler()
#scaled_numerical_data = scaler.fit_transform(numerical_data)
scaled_numerical_data = pd.DataFrame()
scaled_numerical_data['age'] = (numerical_data['age']-18)/(55-18)
scaled_numerical_data['PHQ_sum'] = (numerical_data['PHQ_sum']-X_min)/(X_max_PHQ-X_min)
scaled_numerical_data['GAD_sum'] = (numerical_data['GAD_sum']-X_min)/(X_max_GAD-X_min)
scaled_numerical_data['EHEALS_sum'] = (numerical_data['EHEALS_sum']-X_min)/(X_max_EHEALS-X_min)
scaled_numerical_data['HEALS_sum'] = (numerical_data['HEALS_sum']-X_min)/(X_max_HEALS-X_min)
scaled_numerical_data['CCS_sum'] = (numerical_data['CCS_sum']-X_min)/(X_max_CCS-X_min)
    
    
scaled_numerical_df = pd.DataFrame(scaled_numerical_data, columns=numerical_columns, index=df.index)

# df.filter select columns: match regular expression ^gender_
# ^: anchors the match at the beginning of the string.
# gender_: matches the string "gender_".
df_scaled = pd.concat([scaled_numerical_df, df.filter(regex='^gender_')], axis=1)

print(df_scaled)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fclusterdata

linkageData = linkage(df_scaled,method='average',metric='euclidean')

plt.figure(figsize=(25,25))
dendrogram(linkageData)
plt.show()
```

***Method -> AVERAGE***

```python
from sklearn.metrics import silhouette_samples,silhouette_score

clusters = []
clusterLabels = fclusterdata(df_scaled,t=3,criterion='maxclust',metric='euclidean',method='average')

unique = np.unique(clusterLabels)
print("Checking the number of output clusters.\nOutput classes: ",unique)


silhouette_avg = silhouette_score(df, clusterLabels)
print("For n_clusters =",str(3),"The average silhouette_score is :",silhouette_avg)
sample_silhouette_values = silhouette_samples(df, clusterLabels)
print(sample_silhouette_values)
print("Average silhouette: ", np.mean(sample_silhouette_values))

count = 0
for value in sample_silhouette_values:
    if value <= 0:
        count += 1
print("Number of values below 0: ", count)
```
